```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time  # Importante para las pausas

```

```python
base_url = "https://books.toscrape.com/catalogue/page-{}.html"

datos = []

for page in range(1, 51):
    url = base_url.format(page)
    req = requests.get(url)
    soup = BeautifulSoup(req.text, "lxml")

    titulos = soup.find_all("h3")
    precios = soup.find_all("p", class_="price_color")
    stock = soup.find_all("p", class_="instock availability")
    
    # Transformamos a float aquí o dentro del for, pero no ambos
    precios_limpios = [float(p.text[2:]) for p in precios]

    for item in zip(titulos, precios_limpios, stock):
        titulo = item[0].find("a").get("title")
        precio = item[1]  # <--- Quitamos el .text porque ya es float
        stock= item[2].text.strip()

        datos.append({
            "titulo": titulo,
            "precio": precio, # <--- Usamos la variable 'precio' individual
            "pagina": page,
            "Disponibilidad": stock
        })

df = pd.DataFrame(datos)

print(df)
```
